```python
from transformers import pipeline
```

#2. Setup Generator

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-2.7B') # Second line
```

#3.Generate Text Using Prompt

```python
prompt = "generate a code to print number from 1 to 5" # Third line
```

```python
prompt2 = "Hi!!! How are you?" # Third line
```

```python
res = generator(prompt2, max_length=100, do_sample=True, temperature=0.9) # Fourth line
#do_sample: A boolean parameter that determines whether to sample from the generated probabilities (True) or use the most likely token (False).
#temperature: Controls the randomness of the generated text. Higher values (e.g., 0.9) make the output more diverse and creative, while lower values (e.g., 0.1) make it more deterministic and focused.
```
